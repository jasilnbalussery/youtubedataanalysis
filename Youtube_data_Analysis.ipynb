```python
# ipywidget
import ipywidgets as widgets
```

```python
c=widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Music',
    description='Genre:',
    disabled=False,
)
def topchannel(a):
  channel=video[video['Genre']==a]
  final=channel.groupby(by='channelTitle').sum().sort_values(by='view_count',ascending=False).head()
  sns.barplot(y=final.index,x=final['view_count'],estimator=sum)
  plt.title(a)
  plt.show()
widgets.interact(topchannel,a=c)
```
